```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Science Plotting `(vplot)`

```python
mpl.rc('text', usetex = True, color='k')
mpl.rc('font', family = 'serif')
mpl.rc('figure', figsize=(12,8))
mpl.rc('xtick', labelsize=16, c='k')
mpl.rc('ytick', labelsize=16, c='k')
mpl.rc('axes', ec='k', titlesize=24, labelsize=20, labelcolor='k')
mpl.rc('legend', fontsize=16)
```

# Report Plotting `(vreportplot)`

```python
mpl.rc('text', usetex = True, color='k')
mpl.rc('font', family = 'serif')
mpl.rc('figure', figsize=(8,6))
mpl.rc('xtick', labelsize=20, c='k')
mpl.rc('ytick', labelsize=20, c='k')
mpl.rc('axes', ec='k', titlesize=28, labelsize=24, labelcolor='k')
mpl.rc('legend', fontsize=20)
```

# Presentation Plotting (Black on White) `(vpresplotbow)`

```python
mpl.rc('text', usetex = True, color='k')
mpl.rc('font', family = 'serif')
mpl.rc('figure', figsize=(8, 6))
mpl.rc('xtick', labelsize=18, c='k')
mpl.rc('ytick', labelsize=18, c='k')
mpl.rc('axes', ec='k', titlesize=24, labelsize=20, labelcolor='k')
mpl.rc('legend', fontsize=18)
mpl.rc('savefig', transparent=True)
```

# Presentation Plotting (White on Black) `(vpresplotwob)`

```python
mpl.rc('text', usetex = True, color='w')
mpl.rc('font', family = 'serif')
mpl.rc('figure', figsize=(8, 6))
mpl.rc('xtick', labelsize=18, c='w')
mpl.rc('ytick', labelsize=18, c='w')
mpl.rc('axes', ec='w', titlesize=24, labelsize=20, labelcolor='w')
mpl.rc('legend', fontsize=18, frameon=False)
mpl.rc('savefig', transparent=True)
```

# Poster Plotting `(vposterplot)`

```python
mpl.rc('text', usetex = True, color='k')
mpl.rc('font', family = 'serif')
mpl.rc('figure', figsize=(10, 7.5))
mpl.rc('lines', lw=4, markersize=12)
mpl.rc('xtick', labelsize=24, c='k')
mpl.rc('ytick', labelsize=24, c='k')
mpl.rc('axes', ec='k', titlesize=32, labelsize=28, labelcolor='k')
mpl.rc('legend', fontsize=24)
mpl.rc('savefig', transparent=True)
```

# Sample plots for testing

```python
plt.plot(range(10), label='foo')
plt.plot(range(10), '.', label='bar')
plt.title('foobar')
plt.xlabel('bleh')
plt.ylabel('blah')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.imshow(np.eye(5))
plt.xlabel('wow')
plt.ylabel('nah')
plt.colorbar(label='yee')
plt.tight_layout()
plt.show()
```
